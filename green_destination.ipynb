```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('green_destination_hr.csv')
```

```python
df.head()
```

```python
#Number of employees
df['EmployeeCount'].value_counts()
```

```python
#how many peole have left the company
df['Attrition'].value_counts()
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(
    x='Attrition',
    data=df,
    palette='Set2'
)

plt.title(
    'Employee Attrition Distribution',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel('Attrition Status', fontsize=11)
plt.ylabel('Number of Employees', fontsize=11)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
x = df['Attrition'].value_counts()

plt.figure(figsize=(5, 5))

plt.pie(
    x,
    labels=x.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0.05, 0.05),
    shadow=True
)

plt.title(
    'Employee Attrition Percentage',
    fontsize=14,
    fontweight='bold'
)

plt.axis('equal')  # Makes pie a perfect circle
plt.show()
```
